```python
li = []
for item in data['result']:
    dic = {
        '宗地名称': None,
        'ID': None,
        '建设单位': None,
    }
    for key, value in item.items():
        if key == "建设单位":
            dic['建设单位'] = value
        else:
            dic['宗地名称'] = key
            dic['ID'] = value
    li.append(dic)
```

```python
li = []
for d in data["result"]:
#     print(d)
    msg = {}
    msg['项目名称'] = d['项目名称']
    msg['建设单位(1)'] = d['建设单位(1)']
    msg['建设面积'] = d['建设面积']
    try:
        msg['建设单位'] = d['建设单位']
#     print(msg)

        count = len(d['detail'])
        for i in range(0, count, 2):
            if d['detail'][i] in msg.keys():
                if d['detail'][i+1] not in msg.get(d['detail'][i]):
                    msg[d['detail'][i]] = msg.get(d['detail'][i]) +'/' + d['detail'][i+1]
            else:
                msg[d['detail'][i]] = d['detail'][i+1]
    except:
        pass
    li.append(msg)
```

```python
result = []
for l in li:
    dic = {
        '项目名称': None,
        '建设单位': None,
        '监理企业': None,
        '施工企业': None,
        '设计企业': None,
        "建设单位(1)": None,
    }
    for key,value in l.items():
        if key in dic.keys():
            dic[key] = value
    result.append(dic)
```

```python
import base64
from Crypto.Cipher import AES

class AESDecrypt:
    iv = '0123456789ABCDEF'
    key = 'jo8j9wGw%6HbxfFn'

    @classmethod
    def _pkcs7unpadding(cls, text):
        """
        处理使用PKCS7填充过的数据
        :param text: 解密后的字符串
        :return:
        """
        length = len(text)
        unpadding = ord(text[length-1])
        return text[0:length-unpadding]

    @classmethod
    def decrypt(cls, content):
        """
        AES解密，模式cbc，去填充pkcs7
        :param content: 16进制编码的加密字符串
        :return: 返回解密后的字符串
        """
        key = bytes(cls.key, encoding='utf-8')
        iv = bytes(cls.iv, encoding='utf-8')
        cipher = AES.new(key, AES.MODE_CBC, iv)
        decrypt_bytes = cipher.decrypt(bytes.fromhex(content))
        result = str(decrypt_bytes, encoding='utf-8')
        result = cls._pkcs7unpadding(result)
        return result
```
